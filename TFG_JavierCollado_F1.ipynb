```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt

scaler = MinMaxScaler()

# Adjust labels to start from 0
y_train_adjusted = y_train - 1
y_test_adjusted = y_test - 1

# Initialize lists to store results
best_accuracies = []

# Perform 3 iterations with different random seeds
random_seeds = [42, 123, 456]  # Example random seeds
for iteration, seed in enumerate(random_seeds):
    print(f"\n--- Training Iteration {iteration + 1} with random seed: {seed} ---")
    tf.random.set_seed(seed)  # Set random seed for reproducibility

    # Define the CNN model
    model = keras.Sequential(
        [
            keras.Input(shape=(X_train_s.shape[1], 1)),  # Input shape: (10 features, 1 channel)
            
            # First Conv1D block
            layers.Conv1D(32, kernel_size=3, activation="relu", padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling1D(pool_size=2),
            layers.Dropout(0.3),  # Dropout for regularization
            
            # Second Conv1D block
            layers.Conv1D(64, kernel_size=3, activation="relu", padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling1D(pool_size=2),
            layers.Dropout(0.3),  # Dropout for regularization
            
            # Flatten and Dense layers
            layers.Flatten(),
            layers.Dense(128, activation="relu", kernel_regularizer=keras.regularizers.l2(0.01)),  # L2 regularization
            layers.Dropout(0.3),  # Increased dropout
            layers.Dense(len(np.unique(y_train_adjusted)), activation="softmax"),  # Output layer for classification
        ]
    )

    # Compile the model using ADAM with a reduced learning rate
    model.compile(
        loss="sparse_categorical_crossentropy",
        optimizer=keras.optimizers.Adam(learning_rate=1e-4),  # Reduced learning rate
        metrics=["accuracy"]
    )

    # Reshape input data for CNN
    X_train_cnn = X_train_s.reshape(X_train_s.shape[0], X_train_s.shape[1], 1)
    X_test_cnn = X_test_s.reshape(X_test_s.shape[0], X_test_s.shape[1], 1)

    # Define EarlyStopping callback
    early_stopping = EarlyStopping(
        monitor='val_loss',  # Monitor validation loss
        patience=20,         # Stop if no improvement for 20 epochs
        restore_best_weights=True  # Restore the best weights after stopping
    )

    # Train the model using adjusted labels
    history = model.fit(
        X_train_cnn, 
        y_train_adjusted, 
        epochs=1000, 
        batch_size=512,  # Increased batch size for large datasets
        validation_split=0.3, 
        callbacks=[early_stopping],  # Add EarlyStopping callback
        verbose=1
    )

    # Get the best validation accuracy from this iteration
    best_val_accuracy = max(history.history['val_accuracy'])
    best_accuracies.append(best_val_accuracy)

    # Print the best validation accuracy for this iteration
    print(f"Best Validation Accuracy: {best_val_accuracy:.4f}")

    # Plot training and validation loss
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f"Training and Validation Loss (Iteration {iteration + 1})")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

# Calculate average and standard deviation of the best accuracies
average_best_accuracy = np.mean(best_accuracies)
std_best_accuracy = np.std(best_accuracies)

# Print average and deviation results
print("\n--- Final Results ---")
print(f"Average Best Validation Accuracy: {average_best_accuracy:.4f}")
print(f"Best Validation Accuracy Standard Deviation: {std_best_accuracy:.4f}")
```
